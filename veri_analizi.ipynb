```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("temizlenmis_veri.csv")
```

```python
# İlk 5 kaydı görür
df.head()
```

```python
# Tarih-saat formatına dönüştür
df['Departure'] = pd.to_datetime(df['Departure'], errors='coerce')

# Departure Saati ve Günü Çıkart
df['Departure_Hour'] = df['Departure'].dt.hour
df['Departure_DayOfWeek'] = df['Departure'].dt.dayofweek  # 0 = Pazartesi
```

```python
df.head()
```

```python
# Delay type'ı sayısal hale getir
delay_mapping = {
    'LESS_THAN_30_MINUTES': 1,
    'BETWEEN_30_AND_60_MINUTES': 2,
    'BETWEEN_60_AND_120_MINUTES': 3,
    'OVER_120_MINUTES_OR_CANCELLED': 4
}

df['Delay_Class'] = df['Result'].map(delay_mapping)
```

```python
# Probability Yüzdelik Formata Çevir
df['Probability'] = df['Probability'] * 100
df.head()
```

```python
# Tüm uçuşların ortalama gecikme olasılığı 
print(df['Probability'].mean())
```

```python
# Örneğin gecikme olasılığı %50’den fazla olan uçuşları görmek:
delayed_flights = df[df['Probability'] > 50]
print(delayed_flights.shape)
delayed_flights.head()
```

```python
# Örneğin rota bazlı ortalama gecikme:
avg_delay_by_route = df.groupby('Route')['Probability'].mean().sort_values(ascending=False)
print(avg_delay_by_route)
```

```python
std_delay = np.std(df['Probability'])

print(f"Gecikme Standart Sapması: {std_delay:.2f}%")
```
